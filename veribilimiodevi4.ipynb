```python
#1 numpy ile matris
import numpy as np
arr1=np.random.randint(100,size=(5,5))
print(f"{np.mean(arr1)} ortalaması")
print(f"{np.std(arr1)} standart sapması")
print(f"{np.var(arr1)} varyansı")
print(f"{np.min(arr1)} en küçük sayı")
print(f"{np.max(arr1)} en büyük sayı")
print(f"{np.trace(arr1)} köşegen toplamı")
```

```python
#2 numpy ile veri simülasyonu
np.random.seed(0) # sayıyı sabitlemek için kullandım
puanlar = np.random.normal(loc=70, scale=15, size=1000)
print(f"{np.median(puanlar)} medyan")
print(f"{np.mean(puanlar)} ortalama")
print(f"{np.std(puanlar)} standart sapması")
print(f"{np.sum(puanlar<50)} kişidir 50 den düşük alan öğrenci sayısı")
```

```python
import pandas as pd
bilgi=pd.DataFrame({"öğrenci":["ali","ayşe","mehmet","zeynep","ahmet"],
                    "yaş":[20,21,19,22,20],
                    "bölüm":["bilgisayar","fizik","kimya","bilgisayar","fizik"],
                    "matematik":[70,60,80,90,55],
                    "fizik":[65,75,70,85,60],
                    "kimya":[80,85,65,95,70]}, columns=["öğrenci","yaş","bölüm","matematik","fizik","kimya"],)
bilgi["ortalama"]=bilgi[["matematik","fizik","kimya"]].mean(axis=1) #ortalama sütunu ekledim
print(f"{bilgi["matematik"].mean()} matematik ortalaması")
print(f"{bilgi["matematik"].max()} en yüksek matematik notu")
bolum_ortalama = bilgi.groupby("bölüm")["ortalama"].mean()
print(bolum_ortalama)
print(bilgi[bilgi["ortalama"]<70]) # ortalamasın 70 den büyük olanı filtrelemek için
```
